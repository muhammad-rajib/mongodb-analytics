```python
import certifi
import pprint
from pymongo import MongoClient
import dateparser
```

```python
dateparser.parse("Wednesday, August 12, 2015 at 5:40:37 PM").strftime("%Y-%m-%d, %H:%M:%S")
```

```python
ca = certifi.where()

# Replace XYZ with your connection URI from the Atlas UI
url = 'XYZ'

client = MongoClient(url, tlsCAFile=ca)

client.cleansing.people_raw.update_many( 
    {'birthday': {'$type': 'string'}}, 
    [
        {
            '$set': {
                'birthday': {
                    '$cond': {
                        'if': {'birthday': {'$type': 'string'}},
                        'then': dateparser.parse('$birthday'),
                        'else': '$birthday'
                    }
                }
            }
        }
    ]
)
```
