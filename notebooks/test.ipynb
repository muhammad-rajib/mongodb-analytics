```python
from datetime import datetime
import dateparser
import pymongo
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client['testdb']
db
```

```python
people_with_string_birthdays = db.person.find({'birthday': {'$type': 'string'}})

updates = []

for person in people_with_string_birthdays:
    updates.append(pymongo.UpdateOne({ "_id": person["_id"] }, { '$set': { "birthday": dateparser.parse(person["birthday"]) } }))
        
if updates:         
    db.person.bulk_write(updates)
```

```python
db.person.update_many( 
    {'birthday': {'$type': 'string'}}, 
    [   
        {
            '$set': {'birthday': dateparser.parse('$birthday')}
        }
    ]
)
```
