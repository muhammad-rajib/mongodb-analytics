```python
# import required libraries
import certifi
import pymongo
import pprint
from IPython.display import clear_output
```

```python
# make connection to MongoDB Atlas
ca = certifi.where()

# Replace XYZ with your connection URI from the Atlas UI
url = 'XYZ'

client = pymongo.MongoClient(url, tlsCAFile=ca)
```

```python
# pipline
pipline = [
    # stage 1: GroupBy
    {
        '$sortByCount': '$language'
    },
    # stage 2:
    {
        '$facet':{
            'top language combinations': [{'$limit': 100}],
            'unusual combinations shared by': [
            # inner stage 1
            {
                '$skip': 100
            },
            # inner stage 2
            {
                '$bucketAuto': {
                    'groupBy': '$count',
                    'buckets': 5,
                    'output': {
                        'language combinations': {'$sum': 1}
                    }
                }
            }]
        }
    }
]
```

```python
# run aggregation
clear_output()
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipline)))
```
