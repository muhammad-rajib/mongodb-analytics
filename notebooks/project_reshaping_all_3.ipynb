```python
# import required libraries
import pymongo
import certifi
```

```python
# make connection to MongoDB Atlas (Cloud Database Server)
ca = certifi.where()

# Replace XYZ with your connection URI from the Atlas UI
url = 'XYZ'

client = pymongo.MongoClient(url, tlsCAFile=ca)
```

```python
# pipline for aggregation
pipline = [
#     # stage 1
#     {
#         '$limit': 100
#     },
    
    # stage 2
    {
        '$addFields': 
        {
            'lastupdated': {
                '$arrayElemAt': [
                    {'$split': ['$lastupdated', '.']},
                    0
                ]
            }
        }
    },
    
    # stage 3
    {
        '$project': {
            'title': 1,
            'year': 1,
            'directors': {'$split': ['$directore', ", "]},
            'actors': {'$split': ['$cast', ', ']},
            'writers': {'$split': ['$writer', ', ']},
            'genres': {'$split': ['$genre', ', ']},
            'languages': {'$split': ['$language', ', ']},
            'countries': {'$split': ['$country', ', ']},
            'plot': 1,
            'fullplot': '$fullplot',
            'rated': '$rating',
            'released': 
            {
                '$cond': 
                {
                    'if': {'$ne': ['$released', ""]},
                    'then': {
                        '$dateFromString': {
                            'dateString': '$released'
                        }
                    },
                    'else': ""
                }
            },
            'runtime': 1,
            'poster': 1,
            'imdb': {
                'id': '$imdbID',
                'rating': '$imdbRating',
                'votes': '$imdbVotes'
            },
            'metacritic': 1,
            'award': 1,
            'type': 1,
            'lastUpdated':
            {
                '$cond':
                {
                    'if': {'$ne': ['$lastupdated', ""]},
                    'then': {
                        '$dateFromString': {
                            'dateString': '$lastupdated',
                            'timezone': 'America/New_York'
                        }
                    },
                    'else': ""
                }
            }
        }
    },
    
    # stage 3
    {
        '$out': 'movies_scratch'
    }
]
```

```python
# run aggregate method
client.mflix.movies_initial.aggregate(pipline)

"""
OUTPUT VALUES:
Open your mflix database from MongoDB Compass OR MongoDB Atlas 
to check documents of movies_scratch collection.
"""
```
